# What is statistics, and why is it important?

Statistics is a branch of mathematics that deals with the collection, analysis, interpretation, presentation, and organization of data. It is important because it allows us to make sense of data, draw conclusions, and make informed decisions in various fields, such as science, business, and social sciences.

# What are the two main types of statistics?

The two main types of statistics are descriptive statistics and inferential statistics.

# What are descriptive statistics?

Descriptive statistics are used to summarize and describe the main features of a dataset. This includes measures like mean, median, mode, variance, and standard deviation, as well as graphical representations like histograms and box plots.

# What is inferential statistics?

Inferential statistics are used to draw conclusions and make inferences about a population based on a sample of data. This involves techniques like hypothesis testing and confidence intervals.

# What is sampling in statistics?

Sampling in statistics is the process of selecting a subset of individuals or items from a larger population to collect data. This is often done because it is impractical or impossible to collect data from the entire population.

# What are the different types of sampling methods?

There are various types of sampling methods, including:

*   **Probability sampling:** Simple random sampling, stratified sampling, cluster sampling, systematic sampling.
*   **Non-probability sampling:** Convenience sampling, voluntary response sampling, purposive sampling, snowball sampling.

# What is the difference between random and non-random sampling?

In random sampling, every member of the population has an equal chance of being selected for the sample. This helps to ensure that the sample is representative of the population. In non-random sampling, the selection of individuals is not based on chance, and the sample may not be representative of the population.

# Define and give examples of qualitative and quantitative data.

*   **Qualitative data:** This type of data is descriptive and non-numerical. It describes qualities or characteristics.
    *   Examples: Colors (red, blue, green), types of fruits (apple, banana, orange), opinions (good, bad, neutral).
*   **Quantitative data:** This type of data is numerical and can be measured or counted.
    *   Examples: Height (in inches or centimeters), age (in years), number of students in a class.

# What are the different types of data in statistics?

Data in statistics can be categorized in several ways, including:

*   **Qualitative vs. Quantitative data** (as defined above)
*   **Discrete vs. Continuous data:** Discrete data can only take specific values (e.g., number of cars), while continuous data can take any value within a range (e.g., height).
*   **Nominal, Ordinal, Interval, and Ratio data** (levels of measurement, explained below)

# Explain nominal, ordinal, interval, and ratio levels of measurement.

These are four levels of measurement that describe the nature of the data and the types of statistical analysis that can be performed:

*   **Nominal:** Data at this level are categories without any inherent order.
    *   Examples: Gender (male, female), marital status (single, married, divorced).
*   **Ordinal:** Data at this level are categories with a defined order, but the difference between categories is not meaningful.
    *   Examples: Education level (high school, bachelor's, master's), rankings (1st, 2nd, 3rd).
*   **Interval:** Data at this level have ordered categories with meaningful differences between values, but there is no true zero point.
    *   Examples: Temperature in Celsius or Fahrenheit, years.
*   **Ratio:** Data at this level have ordered categories with meaningful differences and a true zero point, allowing for ratios to be calculated.
    *   Examples: Height, weight, income, age.

# What is the measure of central tendency?

The measure of central tendency is a single value that represents the typical or central value of a dataset. The most common measures are the mean, median, and mode.

# Define mean, median, and mode.

*   **Mean:** The average of a dataset, calculated by summing all values and dividing by the number of values.
*   **Median:** The middle value in a dataset when it is ordered from least to greatest. If there is an even number of values, the median is the average of the two middle values.
*   **Mode:** The value that appears most frequently in a dataset. A dataset can have one mode (unimodal), more than one mode (multimodal), or no mode.

# What is the significance of the measure of central tendency?

The measure of central tendency provides a concise summary of the typical value in a dataset. It helps us understand where the data is centered and allows for comparisons between different datasets.

# What is variance, and how is it calculated?

Variance is a measure of how spread out the data is from the mean. It is calculated by:

1.  Finding the difference between each data point and the mean.
2.  Squaring each of these differences.
3.  Summing the squared differences.
4.  Dividing the sum by the number of data points (for population variance) or by the number of data points minus one (for sample variance).

# What is standard deviation, and why is it important?

Standard deviation is the square root of the variance. It is a widely used measure of the spread of data and is important because:

*   It is in the same units as the original data, making it easier to interpret than variance.
*   It provides a clear indication of how much individual data points typically deviate from the mean.
*   It is used in many statistical formulas and tests.

# Define and explain the term range in statistics.

The range is the simplest measure of spread in a dataset. It is the difference between the highest and lowest values in the dataset.

# What is the difference between variance and standard deviation?

Both variance and standard deviation measure the spread of data, but they are in different units. Variance is in squared units of the original data, while standard deviation is in the same units as the original data. Standard deviation is generally preferred for interpretation because it is easier to understand in the context of the data.

# What is skewness in a dataset?

Skewness is a measure of the asymmetry of the probability distribution of a dataset. It indicates whether the data is skewed to the left (negative skewness) or to the right (positive skewness) relative to the mean.

# What does it mean if a dataset is positively or negatively skewed?

*   **Positively skewed (right-skewed):** The tail of the distribution is longer on the right side, meaning there are more extreme values on the higher end of the data. The mean is typically greater than the median.
*   **Negatively skewed (left-skewed):** The tail of the distribution is longer on the left side, meaning there are more extreme values on the lower end of the data. The mean is typically less than the median.

# Define and explain kurtosis.

Kurtosis is a measure of the "tailedness" of the probability distribution of a dataset. It describes how heavy or light the tails are compared to a normal distribution.

*   **Leptokurtic:** Has heavier tails and a sharper peak than a normal distribution.
*   **Mesokurtic:** Has tails and a peak similar to a normal distribution.
*   **Platykurtic:** Has lighter tails and a flatter peak than a normal distribution.

# What is the purpose of covariance?

Covariance is a measure of the directional relationship between two variables. It indicates whether two variables tend to increase or decrease together. A positive covariance means they tend to move in the same direction, while a negative covariance means they tend to move in opposite directions. However, covariance does not indicate the strength of the relationship.

# What does correlation measure in statistics?

Correlation measures the strength and direction of the linear relationship between two variables. It is a standardized measure that ranges from -1 to +1.

*   A correlation of +1 indicates a perfect positive linear relationship.
*   A correlation of -1 indicates a perfect negative linear relationship.
*   A correlation of 0 indicates no linear relationship.

# What is the difference between covariance and correlation?

The main difference between covariance and correlation is that correlation is a standardized measure, while covariance is not. This means that correlation is not affected by the scale of the variables, making it easier to compare the strength of relationships between different pairs of variables. Covariance is in the units of the product of the two variables, which can make it difficult to interpret.

# What are some real-world applications of statistics?

Statistics has numerous real-world applications, including:

*   **Business:** Market research, quality control, forecasting sales.
*   **Healthcare:** Clinical trials, disease tracking, analyzing patient outcomes.
*   **Finance:** Risk assessment, portfolio management, economic forecasting.
*   **Social Sciences:** Surveys and polls, analyzing demographic trends, studying social behaviors.
*   **Environmental Science:** Analyzing climate data, tracking pollution levels.
*   **Sports:** Analyzing player performance, developing game strategies.

```python
#  [4] Implement sampling techniques like random sampling and stratified sampling
import pandas as pd
from sklearn.model_selection import train_test_split

# Create dataset
df = pd.DataFrame({
    'Gender': ['M', 'F'] * 50,
    'Score': np.random.randint(50, 100, 100)
})

# Random Sampling
random_sample = df.sample(n=10)

# Stratified Sampling
stratified_sample, _ = train_test_split(df, test_size=0.9, stratify=df['Gender'])

print("Random Sample:\n", random_sample)
print("Stratified Sample:\n", stratified_sample)


# The split is done in a stratified way based on the Gender column, meaning:

# The distribution of genders in the 10% sample will match the distribution in the full DataFrame.

# For example, if df['Gender'] is 60% Male and 40% Female, the sample will roughly maintain that same ratio.
```

```python
# Q12. Implement and compare simple random sampling and systematic sampling
import pandas as pd

# Sample dataset
data = pd.DataFrame({'ID': range(1, 11)})

# Simple Random Sampling (pick 4 random rows)
random_sample = data.sample(n=4)
print("Simple Random Sample:\n", random_sample)

# Systematic Sampling (every 2nd row)
systematic_sample = data.iloc[::2]   #iloc stands for "integer location". It is used to select rows and columns by position (not by labels).
print("Systematic Sample:\n", systematic_sample)
```
